```python
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.array([[0,1],[3,4],[6,6],[4,3],[5,5],[2,3],[5,2],[7,3],[9,5],[2,1]])

gmm = GaussianMixture(n_components=2)
gmm.fit(x)
label = gmm.predict(x)
colors = ['.r','.g']
for i in range(len(x)):
  plt.plot(x[i][0],x[i][1],colors[label[i]],markersize=12)

plt.show()
```
