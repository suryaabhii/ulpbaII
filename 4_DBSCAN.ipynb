```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
X = np.array([[2,1],[2,2],[3,2],[7,8],[8,9],[80,25],[75,30]])
X = StandardScaler().fit_transform(X) #Standardization
dbscan = DBSCAN(eps=2, min_samples=3) #epscilon
labels = dbscan.fit_predict(X)        #Data Training
print("DBSCAN Labels:", labels)
colors = ['red', 'green']
for i in range(len(labels)):
  plt.scatter(X[i, 0], X[i, 1], color=colors[labels[i]], s=20)

plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
